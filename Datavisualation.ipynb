# **Line Plot**

Çizgi grafiği, temel olarak iki sayısal değer kümesi arasındaki ilişkiyi göstermek için kullanılır. Genellikle, iki bağımlı d
eği şken
arasında artan veya azalan bir eğilim göstermek için uygundur.

```python
plt.plot(x, y)
plt.xlabel("X Ekseni")
plt.ylabel("Y Ekseni")
plt.show()
```

Burada plot'a x değerlerimizi ve bu değerlere karşılık gelen y değerlerimizi veriyoruz. Ayrıca x ve y eksenlerine label(başlık) veriyoruz. show() metodunu kullanarak oluşturduğumuz grafiği gösteriyoruz.

# **Scatter Plot**

Bir dağılım grafiği, esas olarak iki sayısal grup arasındaki ilişkiyi dağınık noktalar şeklinde çizmek için kullanılır.
Grafikteki her nokta tek bir gözlemi temsil edecek şekilde gösterilir.

Dağılım grafikleri, iki değişken arasındaki ilişkiyi veya
korelasyonu göstermek için kullanışlıdır.

```python
 import numpy as np
```

Random sayılar oluşturmak için numpy kütüphanesinin fonksiyonlarından  yararlanıyoruz.

Oluşturduğumuz random verileri scatter() fonksiyonuna veriyoruz ve grafiğimizi oluşturup gösteriyoruz.

# Pie Chart

Pasta grafikler, adında anlaşılacağı gibi, kategorik bir sütundaki değerlerin yüzde dağılımını gösterir.

* Bir bütünün parçaları arasındaki ilişki açık halde
görülebilir.

* Grafiğin parçaları, her kategorideki bütünün kesri ile
orantılıdır.

```python
etiketler = ["Elma", "Muz", "Kiraz", "Erik"]
sayi = [37.8, 22.2, 20.8 ,19.2]
```

Yine Küçük bir veri seti oluşturuyoruz.

Burada pie chartımızı oluşturuyoruz değerlerimiz sayi labellarımız etiketler, autopct iise virgülden sonra kaç basamak göstereceiğimiz.

# Subplots

Bu özellik, bir figür içinde birden çok alt grafik oluşturmanıza ve bu alt grafiklerin farklı veri setleri veya farklı görüntüleme özellikleri ile oluşturulmasına izin verir.

subplots kullanarak, tek bir figür içinde birden çok alt grafik oluşturabilir ve bu alt grafiklerin düzenini ve yerleşimini belirleyebilirsiniz. Bu şekilde, aynı anda birçok farklı veri veya analiz sonucunu görselleştirebilir ve karşılaştırabilirsiniz.

```python
fig = plt.figure(figsize=(20,5))

first_plot = fig.add_subplot(1,3,1)
first_plot.plot(x, y, color="red")
first_plot.set_title("Line Plot")
first_plot.grid()

second_plot = fig.add_subplot(1,3,2)
second_plot.scatter(x, y, color="green")
second_plot.set_title("Scatter Plot")
second_plot.grid()

third_plot = fig.add_subplot(1,3,3)
third_plot.bar(x, y, color="orange")
third_plot.set_title("Bar Plot")

plt.show()
```

fig = plt.figure(figsize=(20,5)): Bir figür oluşturur ve bu figürün boyutunu (genişlik, yükseklik) belirler. Burada, genişliği 20 birim, yüksekliği 5 birim olarak belirledik.

add_subplot(1,3,1) fonksiyonuyla grafiğimizin konumunu belirliyoruz. Bir satırda (1), üç sütundan oluşan bir düzen içinde (3), ve ilk sütunda (1) yer alır. Bu alt grafik bir çizgi grafiği oluşturduk.
